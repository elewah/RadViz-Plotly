# RadViz-Plotly Demo: Breast Cancer Dataset

This notebook demonstrates the three main visualization functions in RadViz-Plotly:
1. **RadViz2D** - 2D Radial Visualization
2. **RadViz3D** - 3D Radial Visualization
3. **RadViz3D_withSorting** - 3D with intelligent attribute sorting

**Dataset:** Wisconsin Breast Cancer (569 samples, 30 features, 2 classes: Malignant/Benign)

````python
# Import required libraries
import pandas as pd
import plotly.io as pio
from radviz_plotly import RadViz2D, RadViz3D, RadViz3D_withSorting

# Configure Plotly renderer for Colab
pio.renderers.default = 'colab'

print("✓ Libraries imported successfully!")
````

````python
# Load Breast Cancer dataset from GitHub
DATA_URL = "https://raw.githubusercontent.com/elewah/RadViz-Plotly/master/tests/DataFolder/BreastCancer.csv"
data = pd.read_csv(DATA_URL)

print(f"Dataset shape: {data.shape}")
print(f"\nLabel distribution:")
print(data['index'].value_counts())
print(f"\nFirst few rows:")
data.head()
````

````python
# Prepare data for RadViz
y = data['index']  # Labels (M=Malignant, B=Benign)
X = data.drop(['index'], axis=1)  # Features

print(f"Features: {X.shape[1]} attributes")
print(f"Samples: {X.shape[0]} data points")
````

## 1. RadViz2D - 2D Radial Visualization

RadViz2D projects high-dimensional data onto a 2D circle. Each dimension anchor is placed on the circle's perimeter, and data points are positioned based on their normalized feature values, which act as spring forces pulling the point toward each anchor.

**Interactive Features:**
- **Zoom**: Scroll or use toolbar
- **Pan**: Click and drag
- **Filter labels**: Click legend items to show/hide classes
- **Hover**: See data point details
- **Export**: Download as PNG using toolbar

````python
# Create 2D RadViz visualization
RadViz2D(y.copy(), X.copy(), BPs=10000)
````

## 2. RadViz3D - 3D Radial Visualization

RadViz3D extends the concept to 3D by placing dimension anchors on a sphere's surface. This provides additional space for visualizing high-dimensional data with less overlap and better separation between classes.

**Interactive Features:**
- **Rotate**: Click and drag to rotate the 3D view
- **Zoom**: Scroll or use toolbar
- **Pan**: Right-click and drag
- **Filter labels**: Click legend items
- **Reset view**: Double-click the plot

````python
# Create 3D RadViz visualization
RadViz3D(y.copy(), X.copy(), BPs=10000)
````

## 3. RadViz3D_withSorting - Intelligent Attribute Sorting

This advanced version analyzes correlations between attributes and intelligently arranges dimension anchors on the sphere to:
- **Group highly correlated attributes** together
- **Separate negatively correlated attributes** to opposite sides
- **Minimize visual clutter** and improve interpretability

**How it works:**
1. Computes correlation matrix between all features
2. Identifies groups of highly correlated attributes
3. Uses K-means clustering to determine optimal anchor positions
4. Arranges anchors to maximize distance between negatively correlated pairs

**Result:** Clearer patterns and better class separation!

````python
# Create 3D RadViz with intelligent sorting
# Note: This requires scikit-learn and transformations packages (already installed above)
RadViz3D_withSorting(y.copy(), X.copy(), BPs=10000)
````
